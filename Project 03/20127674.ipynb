# Yêu cầu 1a: Sử dụng toàn bộ 10 đặc trưng đề bài cung cấp (2 điểm) 

```python
def exponential(data):
    A = data[:,:-1]
    b = data[:,-1:].reshape(len(data),)
    return np.linalg.pinv(A) @ b
```

```python
# Gọi hàm hồi quy tuyến tính:
def linearRegression(data, articles):
    xPow = exponential(data)        
    xPowData = []
    
    for i in range(len(xPow)):
        row = []
        row.append('x'+ str(i + 1))
        row.append(articles[i])
        row.append(xPow[i])
        xPowData.append(row)
    
    return pd.DataFrame(xPowData, columns=['x', 'Các tính chất', 'Giá trị tương ứng'])
```

```python
linearRegression(data, articles)
```

# Yêu cầu 1b: Xây dựng mô hình sử dụng duy nhất 1 đặc trưng, tìm mô hình cho kết quả tốt nhất (2 điểm)

Lưu ý: khi sử dụng cross-validation, sinh viên cần xáo trộn dữ liệu 1 lần duy nhất và thực hiện trên toàn bộ đặc trưng

```python
def surplusX(train, test):
    xPow = exponential(train)
    sumSurplus = 0
    
    for i in range(len(test)):
        constX = 0
        for j in range(len(xPow)):
            constX += xPow[j] * test[i][j]
        sumSurplus += np.abs(constX - test[i][-1])
        
    return sumSurplus / len(test)
```

```python
def averageSurplusX(data, num):
    n = int(np.round(len(data) / num))
    models = [data[i*n:i*n+n, :] for i in range(num-1)]
    models.append(data[(num-1)*n:, :])

    surplusNum = []
    for i in range(num):
        
        train = np.concatenate(models[:-1], axis=0)
        test = models[-1]
        surplusNum.append(surplusX(train, test))
        
        temp = models[-1]
        models[-1] = models[i]
        models[i] = temp
    
    return np.average(surplusNum), surplusNum
```

```python
def surplusAverageAttribute(data, num):
    surplusNumAverage = []
    surplusNum = []
    
    for i in range(len(data[0]) - 1):
        dataAttribute = np.concatenate((data[:, i:i+1], data[:, -1:]), axis=1)
        surplusAvgAttribute, surplusAttribute = averageSurplusX(dataAttribute, num)
        surplusNumAverage.append(surplusAvgAttribute)
        surplusNum.append(surplusAttribute)
    
    return surplusNumAverage, surplusNum
```

```python
def rankTable(data, articles):
    surplusNumAverage, surplusNums = surplusAverageAttribute(data, 4)
    indices = np.argsort(surplusNumAverage)
    
    rankIn4 = []
    for i in range(len(indices)):
        row = []
        row.append(articles[indices[i]])
        row.append(surplusNums[indices[i]][0])
        row.append(surplusNums[indices[i]][1])
        row.append(surplusNums[indices[i]][2])
        row.append(surplusNums[indices[i]][3])
        row.append(surplusNumAverage[indices[i]])
        row.append(i + 1)
        rankIn4.append(row)
        
    return pd.DataFrame(rankIn4, columns=["Các tính chất", "Sai số 1", "Sai số 2", "Sai số 3", "Sai số 4", "Sai số trung bình", "Rank"])
```

```python
def bestDataTable(data, articles):
    surplusNumAverage, _ = surplusAverageAttribute(data, 4)
    bestRank = np.argsort(surplusNumAverage)[0]
    
    newIn4 = np.concatenate((data[:, bestRank:bestRank+1], data[:,-1:]), axis=1)
    newArticle = articles[bestRank:bestRank+1] + articles[-1:]
    
    return newIn4, newArticle
```

```python
rankTable(data, articles)
```

```python
# Kết quả của câu b.
newIn4, newArticle = bestDataTable(data, articles)
linearRegression(newIn4, newArticle)
```
